# Восстановление золота из руды

**Цель проекта:** построить модель предсказывающая коэффициент восстановления золота из золотосодержащей руды с помошью параметров добычи и очистки. 

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

**Задачи проекта:**
1. Подготовить данные;
2. Провести исследовательский анализ данных;
3. Построить и обучить модель.

**Наименование признаков:**

Наименование признаков имеет следующий вид:

`[этап].[тип_параметра].[название_параметра]`

Пример: `rougher.input.feed_ag`

Возможные значения для блока `[этап]:`
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики

Возможные значения для блока `[тип_параметра]:`
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

**Описание данных**

Технологический процесс:

* Rougher feed — исходное сырье
* Rougher additions (или reagent additions) — флотационные реагенты (Xanthate, Sulphate, Depressant)
    * Xanthate — ксантогенат (промотер, или активатор флотации);
    * Sulphate — сульфат (на данном производстве сульфид натрия);
    * Depressant — депрессант (силикат натрия).
* Rougher process (англ. «грубый процесс») — флотация
* Rougher tails — отвальные хвосты
* Float banks — флотационная установка
* Cleaner process — очистка
* Rougher Au — черновой концентрат золота
* Final Au — финальный концентрат золота

Параметры этапов

* air amount — объём воздуха
* fluid levels — уровень жидкости
* feed size — размер гранул сырья
* feed rate — скорость подачи

### Проверка расчётов

Эффективности обогащения золота на этапе флотации рассчитывается по формуле: $$\cfrac{C * (F-T)}{F * (C-T)}*100\%$$, где C — доля золота в концентрате после флотации/очистки, F — доля золота в сырье/концентрате до флотации/очистки, T — доля золота в отвальных хвостах после флотации/очистки.

```python
C = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.concentrate_au']
F = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.input.feed_au']
T = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.tail_au']
rough_recov_calc = C * (F - T) / (F * (C - T)) * 100
rough_recov_train = train.loc[~train['rougher.output.recovery'].isna(), 'rougher.output.recovery']
```

MAE между расчётными значениями и значением признака рассчитывается по формуле: $$sMAPE = \cfrac{1}{N}*\sum_{i=1}^n \cfrac{|y_i - \hat{y_i}|}{\left(\cfrac{|y_i| - |\hat{y_i}|}{2} \right)}*100\%$$, где N - количество объектов в выборке, $\sum_{i=1}^n$ - суммирование по всем объектам выборки (i меняется от 1 до N), $y_i$ - значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество, $\hat{y_i}$ - значение предсказания для объекта с порядковым номером i, например, в тестовой выборке.

```python
# расчёт sMAPE 
smape_rough_rec = (1 / len(rough_recov_train) * (abs(rough_recov_train - rough_recov_calc) 
                                                 / ((abs(rough_recov_train) + abs(rough_recov_calc))/2)).sum() * 100)
print("sMAPE: {0:.2E}".format(smape_rough_rec))
```

Значение ошибки sMAPE имеет очень малый процент, следовательно, эффективность рассчитана верно.

### Функция вычисления итоговой sMAPE

```python
def smape(target, predict): 
    return (1 / len(target) * (abs(target - predict) / ((abs(target) + abs(predict))/2)).sum() * 100)
```

```python
def smape_final(smape_rougher, smape_final):
    return 0.25*smape_rougher + 0.75*smape_final
```

### Обучение и оценка качества различных моделей кросс-валидацией

Для предсказания эффективности обогащения на этапе флотации могут быть использованы только признаки с префиксом rougher.

```python
rougher_columns = []
pattern = re.compile('rougher')
print('Признаки для предсказания эффективности обогащения на этапе флотации:')
for column in list(test.columns):
    if re.match(pattern, column):
        rougher_columns.append(column)
        print(column)
```

Выделим в обучающей выборке признаки и целевые признаки

```python
X_r = train_r[rougher_columns]
X_f = train_f[list(test.columns)]
y_r = train_r['rougher.output.recovery']
y_f = train_f['final.output.recovery']
test_r = test[rougher_columns]
test_f = test[list(test.columns)]
```

Масштабируем признаки на обучающей и тестовой выборках 

```python
ss_r = StandardScaler()
ss_f = StandardScaler()
ss_r.fit(X_r)
ss_f.fit(X_f)

X_r = pd.DataFrame(ss_r.transform(X_r), columns = rougher_columns)
X_f = pd.DataFrame(ss_f.transform(X_f), columns = list(test.columns))
test_f_ind = test.index
test_r_ind = test_r.index
test_r = pd.DataFrame(ss_r.transform(test_r), columns = rougher_columns, index = test_r_ind)
test_f = pd.DataFrame(ss_f.transform(test), columns = list(test.columns), index = test_f_ind)
```

Модель линейной регрессии

```python
%%time
# Модель для этапа флотации
lr_rougher = LinearRegression()
smape_lr_rougher = cross_val_score(lr_rougher, X_r, y_r, scoring = make_scorer(smape), cv = 5, n_jobs = -1).mean()
```

```python
%%time
# Модель для финального этапа 
lr_final = LinearRegression()
smape_lr_final = cross_val_score(lr_final, X_f, y_f, scoring = make_scorer(smape), cv = 5, n_jobs = -1).mean()
```

```python
smape_f_lr = smape_final(smape_lr_rougher, smape_lr_final)
print('Линейная регрессия:')
print('sMAPE rougher: {:.2f} %'.format(smape_lr_rougher))
print('sMAPE cleaner: {:.2f} %'.format(smape_lr_final))
print('Final sMAPE: {:.2f} %'.format(smape_f_lr))
```

Модель RandomForestRegressor

Подберём гиперпараметры для модели с помощью техники GridSearchCV. Она последовательно обучает модель на всех заданных комбинациях гиперпараметров, используя кросс валидацию. Далее возвращает комбинацию гиперпараметров, с которыми модель достигла наилучшего значения метрики качества.

```python
max_depth = range(5,10)
n_estimators = range(5, 10)
param_grid = {'n_estimators':n_estimators, 'max_depth': max_depth}

rf_rougher = RandomForestRegressor()

grid_search_r = GridSearchCV(estimator = rf_rougher, param_grid = param_grid, cv = 5, n_jobs = -1, verbose=1,
                           scoring = make_scorer(smape))
grid_search_r.fit(X_r, y_r)
print('sMAPE rougher: {:.2f} %'.format(grid_search_r.best_score_))
print('Лучшие параметры',grid_search_r.best_params_)
```

```python
rf_final = RandomForestRegressor()

grid_search_f = GridSearchCV(estimator = rf_final, param_grid = param_grid, cv = 5, n_jobs = -1, verbose=1,
                           scoring = make_scorer(smape))
grid_search_f.fit(X_f, y_f)
print('sMAPE rougher: {:.2f} %'.format(grid_search_f.best_score_))
print('Лучшие параметры',grid_search_f.best_params_)
```

```python
smape_f_rf = 0.25*grid_search_r.best_score_ + 0.75*grid_search_f.best_score_
print('Final sMAPE: {:.2f} %'.format(smape_f_rf))
```

Модель Лассо

```python
%%time
las_rougher = Lasso()
smape_las_rougher = cross_val_score(las_rougher, X_r, y_r,scoring = make_scorer(smape), cv = 5, n_jobs = -1).mean()
```

```python
%%time
las_final = Lasso()
smape_las_final = cross_val_score(las_final, X_f, y_f, scoring = make_scorer(smape), cv = 5, n_jobs = -1).mean()
```

```python
smape_f_las = smape_final(smape_las_rougher, smape_las_final)

print('Модель Lasso:')
print('sMAPE rougher: {:.2f} %'.format(smape_las_rougher))
print('sMAPE cleaner: {:.2f} %'.format(smape_las_final))
print('Final sMAPE: {:.2f} %'.format(smape_f_las))
```

Расчеты sMAPE для моделей линейной регрессии и лассо показали, что лучшей моделью является модель лассо. Проверим данную модель на тестовой выборке.

### Тестирование моделей

```python
# из исходных данных сохраним целевые признаки для тестовой выборки
y_true_r = full.loc[test.index, 'rougher.output.recovery'].dropna()
y_true_f = full.loc[test.index, 'final.output.recovery'].dropna()
# создадим тестовые выборки признаков с валидными индексами для обоих этапов
test_r = test_r.loc[y_true_r.index]
test_f = test_f.loc[y_true_f.index]
test_r = test_r[X_r.columns]
test_f = test_f[X_f.columns]
```

```python
model_r = Lasso()
model_f = Lasso()

model_r.fit(X_r, y_r)
model_f.fit(X_f, y_f)
predict_r = model_r.predict(test_r)
predict_f = model_f.predict(test_f)

smape_test_r = smape(y_true_r, predict_r)
smape_test_f = smape(y_true_f, predict_f)

print('sMAPE флотации на тестовой выборке: {:.2f} %'.format(smape_test_r))
print('sMAPE очистки на тестовой выборке:  {:.2f} %'.format(smape_test_f))

smape_test_fin = smape_final(smape_test_r, smape_test_f)
print('Итоговая sMAPE на тестовой выборке: {:.2f} %'.format(smape_test_fin))
```
